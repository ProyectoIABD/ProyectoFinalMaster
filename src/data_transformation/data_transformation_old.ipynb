```python
import pandas as pd
import re
```

```python
tesla = pd.read_csv('../../data/tesla.com.csv')
milanuncios = pd.read_csv('../../data/milanuncios.com.csv')
cochesNet = pd.read_csv('../../data/cochesNet.csv')

tesla.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
milanuncios.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
cochesNet.drop(columns=['Ubicación', 'Tipo de vehículo', 'Link'], axis=1, inplace=True)
```

```python
used = pd.concat([cochesNet, milanuncios], ignore_index=True)
used['sell_type'] = 'segunda_mano'
```

```python
used
```

```python
def extract_model(car_type):
    import re
    # Expresion para detectar "Model 3", "Model Y", "Model S" o "Model X"
    match = re.search(r"Model (3|Y|S|X)", car_type, re.IGNORECASE)
    return f"Model {match.group(1)}" if match else None
```

```python
used["car_type"] = used["car_type"].apply(extract_model)
used.dropna()
```

```python
full_data = pd.concat([tesla, used], ignore_index=True)
full_data
```

```python
full_data.isna().sum()
```

```python
# Dropear filas con valores nulos en car_type y price
full_data.dropna(subset=["car_type", "price"], inplace=True)

# Limpiar y convertir la columna 'km'
full_data['km'] = full_data['km'].str.extract(r'(\d+)').astype(float)

# Limpiar y convertir la columna 'price'
full_data['price'] = (
    full_data['price']
    .str.replace('€', '', regex=False)
    .str.replace(r'[^\d]', '', regex=True)
    .astype(float)
)
```

```python
full_data.isna().sum()
```

```python
full_data.drop(columns=['color', 'country', 'traction', 'range', 'max_vel', 'zero_hundred'], axis=1, inplace=True)
```

```python
full_data.to_csv('../../data/full_data.csv')
```
