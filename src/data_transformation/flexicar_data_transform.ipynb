```python
import pandas as pd
from datetime import datetime
```

```python
#Dataframe
df = pd.read_csv("../../data/raw/flexicar.es.csv")
df
```

```python
top_30_coches = df['name'].value_counts().nlargest(30).index
top_30_coches
```

```python
df_top30 = df[df['name'].isin(top_30_coches)].copy()
df_top30
```

```python
def car_status(year, km):
    current_year = datetime.now().year
    age = max(current_year - year, 1)
    km_per_year = km / age
    km_per_year
    return 'neutral' if km_per_year < 40000 else 'malo'
```

```python
df_top30['km'] = df_top30['km'].str.replace('.', '', regex=False)  # elimina los puntos de miles
df_top30['km'] = df_top30['km'].str.replace(' km', '', regex=False)  # elimina ' km'
df_top30['km'] = df_top30['km'].astype(int)

df_top30["prediccion"] = df_top30.apply(lambda row: car_status(row["year"], row["km"]), axis=1)
```

```python
df_top30.drop(columns=['motor'], inplace=True)
df_top30['prediccion'].value_counts()
```

```python
df_top30.to_csv("../../data/transformed/flexicar.es.csv", index=False)
```
