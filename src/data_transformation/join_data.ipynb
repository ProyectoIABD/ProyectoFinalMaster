```python
import pandas as pd
```

```python
df_flexi = pd.read_csv("../../data/transformed/flexicar.es.csv")
df_mil = pd.read_csv("../../data/transformed/milanuncios_predi.csv")
df_mil
```

```python
df_total = pd.concat([df_flexi, df_mil], ignore_index=True)
df_total
```

```python
df_total = df_total[df_total['car_type'] != 'BMW - SERIE']
```

```python
df_total['car_type'] = df_total['car_type'].str.upper()
```

```python
df_total['car_type'].unique()
```

```python
marcas = ['BMW', 'AUDI', 'TESLA', 'PORSCHE', 'CUPRA']

def formatear_car_type(valor):
    if pd.isna(valor):
        return valor
    valor = str(valor).upper().strip()
    for marca in marcas:
        if valor.startswith(marca):
            resto = valor[len(marca):].strip()
            # Si ya tiene el guion, no lo agregamos
            if resto.startswith('-'):
                resto = resto[1:].strip()
            return f"{marca} - {resto}"
    return valor  # Si no detectamos ninguna marca, lo dejamos igual

# Aplicamos al dataframe
df_total['car_type'] = df_total['car_type'].apply(formatear_car_type)
```

```python
modelos_equivalentes = {
    'AUDI - Q3 SPORTBACK': 'AUDI - Q3',
    'AUDI - Q5 SPORTBACK': 'AUDI - Q5',
    'BMW - SERIE 2 GRAN TOURER': 'BMW - SERIE 2',
    'BMW - SERIE 2 ACTIVE TOURER': 'BMW - SERIE 2',
}

df_total['car_type'] = df_total['car_type'].replace(modelos_equivalentes)
```

```python
df_total.to_csv("../../data/processed/full_data.csv", index=False)
```
